# **Distribuciones**

La distribución de probabilidad de una variable aleatoria es una función que asigna a cada suceso definido sobre la variable la probabilidad de que dicho suceso ocurra. 

**Distribución normal**

La distribución normal es descrita completamente por los dos parámetros mu (µ), la media, y sigma (σ), la desviación estándar, y su media, moda y mediana son equivalentes.  

También se conoce como distribución de Gauss (Gaussian distribution) y su gráfico dibuja una forma acampanada conocida como la campana de Gauss.

La distibución normal estándar es aquella que tiene una media de 0 y una desviación estándar de 1.


**Distribución uniforme**

Una distribución uniforme es una en la que todos los valores (entre el valor máximo y el mínimo) tienen la misma probabilidad de ocurrir, es decir, la probabilidad se distribuye uniformemente a lo largo de un intervalo.

# **Distribución normal vs. uniforme**



```
import torch
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
sns.set()
import warnings
warnings.simplefilter("ignore", UserWarning)
```

```
# Creamos un tensor de números aleatorios con una distribución uniforme
uniform = torch.empty(1000000).uniform_(-4, 4) # Límites inferior y superior de la distribución

# Creamos un tensor de números aleatorios con una distribución normal
normal = torch.empty(1000000).normal_(mean=0,std=1) # Media y desviación estándar de la distribución
```

```
# Graficamos las dos distribuciones
figure(figsize=(10, 8))
ax = sns.distplot(uniform, label='Distribución uniforme', color="purple")
bx = sns.distplot(normal, label= 'Distribución normal', color = "green")
legend = plt.legend()
plt.axvline(color="black")
plt.axvline(x=1, color="black", linestyle=":")
plt.axvline(x=-1, color="black", linestyle=":")
plt.show()
```

**Otras distribuciones normales**

```
normal_1 = torch.empty(10000).normal_(mean=0,std=0.2) 
normal_2 = torch.empty(10000).normal_(mean=0,std=5.0)
normal_3 = torch.empty(10000).normal_(mean=-2,std=0.5)
normal_standard = torch.empty(10000).normal_(mean=0,std=1.0)

figure(figsize=(12, 10))
ax = sns.distplot(normal_1, label='Distribución normal 1', color="purple", hist=False)
bx = sns.distplot(normal_2, label= 'Distribución normal 2', color = "green", hist=False)
cx = sns.distplot(normal_3, label= 'Distribución normal 3', color = "blue", hist=False)
dx = sns.distplot(normal_standard, label= 'Distribución normal estándar', color = "red", hist=False)
legend = plt.legend()
plt.axvline(color="black")
plt.axvline(x=1, color="black", linestyle=":")
plt.show()
```
